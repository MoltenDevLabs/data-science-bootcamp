Para hacer graficos hay que importar matplotlib.pyplot

Para mostrar un grafico hay que usar la funcion show() de pyplot
Para guardar una imagen de un grafico hay que llamar a la funcion savefig() de pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cx = [1,2,3,4]    # Puntos a representar en el eje x
cy = [1,2,0,0.5]  # Puntos a represnetar en el eje y
```

```python
fig, ax = plt.subplots()
ax.plot(cx, cy)   # Tipo de grafico (linea)
plt.show()  # Funcion para crear el grafico
```

```python
fig, ax = plt.subplots()
ax.scatter(cx, cy)  # Tipo de gracico (scatter)
plt.show()  # Funcion para crear el grafico
```

```python
fig, ax = plt.subplots()
ax.fill_between(cx, cy) # Tipo de grafico que pinta el area debajo de la linea
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(cx, cy) # Tipo de grafico de barras
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(cx, cy) # Tipo de grafico de barras horizontales
plt.show()
```


Para definir un histograma lo haremos con la funcion hist() donde los argumentos son la frequencia y el numero de bins

bin -> intervalos

```python
frequencia = np.random.normal(5, 1.6, size=1500)
numero_bins = np.arange(0, 11)  # Rango de 10 valores equidistantes

fig, ax = plt.subplots()
ax.hist(frequencia, numero_bins) # Tipo de grafico histograma
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(frequencia, numero_bins) # Tipo de grafico histograma
plt.show()
```

Grafico de quesito (pie)

```python
a = [1, 1, 1, 1, 1] # Este vector define el cuantas partes hay y el porcentaje de cada parte

fig, ax = plt.subplots()
ax.pie(a) # Tipo de grafico quesito
plt.show()
```

```python
e = [34, 36, 37, 41, 21, 51, 34, 32, 52]  # Lista con distintas edades
print(f"Media: np.mean(e) = {np.mean(e)}")
print(f"Mediana: np.median(e) = {np.median(e)}")
print(f"Maximo: np.max(e) = {np.max(e)}")
print(f"Minimo: np.min(e) = {np.min(e)}")

fig, ax = plt.subplots()
ax.boxplot(e) # Tipo de grafico velas (barras y bigotes)
plt.show()
```

Diagrama de contorno (Ej. uso: Sistemas topolÃ³gicos)

```python
fig, ax = plt.subplots()
x = np.linspace(-7.5, 3, 100) # 100 valores equidistantes entre -7.5 y 3
y = np.linspace(7.5, -6.43, 100) # 100 valores equidistantes entre 7.5 y -6.43
x, y = np.meshgrid(x, y)  # Crea una malla
z = np.sqrt(x**2+y**2) 
ax.contour(x, y, z) # Tipo de grafico diagrama de contorno
plt.show()
```

```python
fig, ax = plt.subplots()
ax.imshow(np.random.random((32, 32))) # Tipo de grafico Â¿diagrama de relaciÃ³n binaria?
plt.show()
```

Ejemplo practico:
meses y ventas de casas

Para hacer graficos lo mejor es ir a la libreria de matplotlib, cojer un grafico ya hecho y modificarlo

https://matplotlib.org/stable/plot_types/index

https://matplotlib.org/cheatsheets/

```python
meses = ["ene", "feb", "mar", "abr", "may", "jun"]
ventas = {"nueva": [100,105,300,267,165,777], "segunda": [50,88,103,200, 20, 155]}

fig, ax = plt.subplots()
ax.plot(meses, ventas["nueva"], color = 'tab:red', marker = '*', linestyle = 'dashed', label = 'Nueva')
ax.plot(meses, ventas["segunda"], color = 'tab:green', marker = '^', linestyle = 'dotted', label = 'Segunda mano')
ax.set_title('EvoluciÃ³n de viviendas en primer semestre', loc = 'left', fontdict = {'fontsize': 8, 'color': 'blue', 'fontweight': 'bold'})
ax.set_xlabel('meses')
ax.set_ylabel('ventas')
ax.set_ylim([10, 1000]) # Rango entre 10 y 1000, ignora el resto de valores
ax.set_yticks(range(10, 1000, 50))
ax.legend(loc = 'upper left') # Inserta la leyenda
ax.grid(axis = 'x', color = 'tab:gray', linestyle = 'dashed')
plt.show()
```

Ejemplo de multiples graficos

```python
# Conjunto de datos a graficar
meses = ["ene", "feb", "mar", "abr", "may", "jun"]
ventas = {"nueva22": [100,105,300,267,165,777], "segunda22": [50,88,103,200, 20, 155], "nueva23": [150,155,390,167,185,310], "segunda23": [100,188,153,20,200,177]}

# ConstrucciÃ³n del conjunto de grÃ¡ficos
fig, ax = plt.subplots(2, 2, sharey=True)

# AÃ±adir los conjuntos de datos
ax[0,0].plot(meses, ventas["nueva22"], color = 'tab:red', marker = '*', linestyle = 'dashed')
ax[0,1].plot(meses, ventas["segunda22"], color = 'tab:green', marker = '^', linestyle = 'dotted')
ax[1,0].plot(meses, ventas["nueva23"], color = 'tab:red', marker = '*', linestyle = 'dashed')
ax[1,1].plot(meses, ventas["segunda23"], color = 'tab:green', marker = '^', linestyle = 'dotted')
# AÃ±adir etiquetas (se usa fig para aÃ±adir etiquetas al conjunto de grÃ¡ficos, mientras que ax para un solo grÃ¡fico)
fig.suptitle('Comparativa ventas 2022 vs 2023', fontweight = 'bold') # Titulo en la parte superior
fig.text(0.5, 0, 'Primer semestre', ha='center')  # 0.5 implica centrado en eje x, 0 implica en la posiciÃ³n 0 del eje y (abajo del todo), ha es para horizontal alignment
fig.text(0, 0.5, 'Numero de ventas por 1000 habitantes', va='center', rotation='vertical')  # 0 implica a la izquierda del eje x, 0.5 implica centrado en el eje y, hv es para vertical alignment, la rotaciÃ³n vertical es para rotar el texto
fig.text(0.04, 0.6, '2022', va='center', rotation='vertical')
fig.text(0.04, 0.4, '2023', va='center', rotation='vertical')
# AÃ±adir leyenda
fig.legend(labels = ["Nueva","Segunda"], bbox_to_anchor=(0.9, 0.88)) # La posiciÃ³n (0,0) corresponde a abajo-izquierda, la (1,1) a arriba-derecha
# Quitar etiquetas de los meses
ax[0,0].set_xticklabels([]) # GrÃ¡fico[0,0], desactivar etiquetas 
ax[0,1].set_xticklabels([]) # GrÃ¡fico[0,1], desactivar etiquetas 
# Quitar ticks
ax[0,0].tick_params(bottom=False) # GrÃ¡fico[0,0], desactivar ticks abajo
ax[0,1].tick_params(left=False, bottom=False) # GrÃ¡fico[0,1], desactivar ticks en la izquierda y abajo
ax[1,1].tick_params(left=False) # GrÃ¡fico[1,1], desactivar ticks en la izquierda

plt.show()
```
